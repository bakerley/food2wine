```python
import string
from nltk.corpus import stopwords 
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
def preprocessing(text):
    for punctuation in string.punctuation: 
        text = text.replace(punctuation, ' ') 
    lowercased = text.lower()
    words_only = ''.join([i for i in lowercased if not i.isdigit()])
    return words_only
    
df['clean_description'] = df.description.apply(preprocessing)
df.head()
```

```python
# remove duplicate
len(df)
df = df.drop_duplicates() # Remove duplicates
len(df)
```

## Stopwords

```python
from nltk.corpus import stopwords 
from nltk import word_tokenize

stop_words = set(stopwords.words('english')) 

# Create function
def remove_stopwords (text):
    tokenized = word_tokenize(text)
    without_stopwords = [word for word in tokenized if not word in stop_words]
    return without_stopwords

df['clean_description'] = df.clean_description.apply(remove_stopwords)

df.head()
```

## Lemmatize

```python
from nltk.stem import WordNetLemmatizer

def lemma(text):
    lemmatizer = WordNetLemmatizer() # Initiate lemmatizer
    lemmatized = [lemmatizer.lemmatize(word) for word in text] # Lemmatize
    lemmatized_string = " ".join(lemmatized)
    return lemmatized_string

df['clean_description'] = df.clean_description.apply(lemma)

df.head()
```
